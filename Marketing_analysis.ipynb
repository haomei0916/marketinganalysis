```python
#convert date to datetime
marketing_df.columns = marketing_df.columns.str.replace(' ', '')
marketing_df['Dt_Customer']=pd.to_datetime(marketing_df['Dt_Customer'])
import warnings
warnings.filterwarnings('ignore')
```

```python
#delating the $ sign from income
marketing_df['Income'] = marketing_df['Income'].astype(str)

marketing_df['Income'] = marketing_df['Income'].str.replace('$', '').str.replace(',', '')
```

```python
# Replace '2n cycle' with 'master'
marketing_df['Education'] = marketing_df['Education'].apply(lambda x: 'Master' if str(x) == '2n Cycle' else str(x))
# Replace 'graduation' with 'undergraduate'
marketing_df['Education'] = marketing_df['Education'].apply(lambda x: 'Undergraduate' if str(x) == 'Graduation' else str(x))
```

```python
# pivot table for see maritalstatus according to education
df1 = marketing_df[["Education", 'Marital_Status']].value_counts().groupby(["Education", "Marital_Status"]).sum().reset_index().sort_values(by=['Education','Marital_Status'], ascending=False)
df1 = (df1.pivot(index='Education', columns='Marital_Status'))
df1
```

```python
#Replacing Missing values based on Education and Marital Status
unique_Education = pd.unique(marketing_df.Education)
unique_MStatus=pd.unique(marketing_df.Marital_Status)
temp_data = marketing_df.copy()  # set temp_data variable to avoid losing real data

all_data_mean = np.round(np.mean(temp_data['Income']), 2)

for q in unique_Education:
    for m in unique_MStatus:
        filter = (temp_data.Education == q) & ( temp_data.Marital_Status == m)
        filtered_data = temp_data[filter]
        
        mean = np.round(np.mean(filtered_data['Income']), 2)
        
        if ~np.isnan(mean): 
            filtered_data['Income'] = filtered_data['Income'].fillna(mean)
            print(f"Missing Value in 'Income' column filled with {mean} when Education:{q} and Marital Status: {m}")
        else: 
            filtered_data['Income'] = filtered_data['Income'].fillna(all_data_mean)
            print(f"Missing Value in 'Income' column filled with {all_data_mean} when Education:{q} and Marital Status: {m}")
        
        temp_data[filter] = filtered_data
    
df = temp_data.copy() 
```

## Hypothesis Testing

1. Older people are not as tech-savvy and probably prefer shopping in-store.

  H0 = Older people are equally as tech-savvy and likely to shop online as younger people.

  H1 = Older people are less tech-savvy and more likely to prefer shopping in-store compared to younger people.

```python
marketing_df_age=marketing_df[['Age','NumWebPurchases','NumStorePurchases']]


marketing_df_age.loc[(marketing_df_age['Age'] <= 25) , 'Age_Group'] = 'Young'
marketing_df_age.loc[(marketing_df_age['Age']  > 50) , 'Age_Group'] = 'Elderly'
marketing_df_age.loc[(marketing_df_age['Age']  > 25) & (marketing_df_age['Age'] <= 50) , 'Age_Group'] = 'Middle Age'


marketing_df_age.head()
```

```python
marketing_df_age_plot=marketing_df_age[["Age_Group", 'NumWebPurchases','NumStorePurchases']].groupby(["Age_Group"]).sum() # Age_Group as an index for plotting .reset_index() 
marketing_df_age_plot.head()
```

```python
marketing_df_age_plot.plot.bar()
```

Age group with people who are 50 years old and up made more purchases online and in-store than middle age group, therefore, we accept null hypothesis and reject alternative hypothesis.

2. Customers with kids probably have less time to visit a store and would prefer to shop online.

H0: The mean preference for online shopping among customers with kids is equal to the mean preference for online shopping among customers without kids.
:

H1: Customers with kids have a different preference for online shopping compared to customers without kid






```python
marketing_df_kids=marketing_df[['Kidhome','NumWebPurchases','NumStorePurchases']]
marketing_df_kids_plot= marketing_df_kids.groupby(["Kidhome"]).sum() #.reset_index() 
marketing_df_kids_plot.head()
```

```python
marketing_df_kids_plot.plot.bar()
```

The ratio of web purchase is bigger as the number of kids becomes bigger, so we reject the H0 and accept H1.

3. Other distribution channels may cannibalize sales at the store.

H0 = There is no effect of other distribution channels on sales at the store.

H1 = Other distribution channels have an effect on sales at the store.

```python
labels = ['Deals','Online','Catalog','Instore']
sizes = [marketing_df.NumDealsPurchases.sum(),
         marketing_df.NumWebPurchases.sum(),
         marketing_df.NumCatalogPurchases.sum(),
         marketing_df.NumStorePurchases.sum()]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

Instore sale is bigger than other channels, but when the channels other than instore are combined, they are bigger than instore. We need more information to conclude this inference.

4. Does the US fare significantly better than the rest of the world in terms of total purchases?

   H0 = The total purchases in the US is equal to the total purchases in the rest of the world.

   H1 = The total purchases in the US is significantly different from the total purchases in the rest of the world.



```python
totalpurchases = marketing_df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'Country']]
totalpurchases.head()
```

```python
# Calculate the total spending by each customer
marketing_df['Total_purchases'] = marketing_df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)
```

```python
# Calculate the total number of purchases made through different channels
totalpurchases['TotalPurchases'] = totalpurchases['NumDealsPurchases'] + totalpurchases['NumWebPurchases'] + totalpurchases['NumCatalogPurchases'] + totalpurchases['NumStorePurchases']
totalpurchases.head()
```

```python
average_purchases_per_country = totalpurchases.groupby('Country').agg(Total_Purchases = ('TotalPurchases', 'sum'))
average_purchases_per_country.head(1)
```

```python
average_purchases_per_country['Total_Customers'] = totalpurchases['Country'].value_counts()
average_purchases_per_country.head(1)
```

```python
average_purchases_per_country['Purchases_Per_Customer'] = np.floor(average_purchases_per_country['Total_Purchases'] / average_purchases_per_country['Total_Customers'])
average_purchases_per_country.head(1)
```

```python
average_purchases_per_country.reset_index(inplace = True)
average_purchases_per_country.head(1)
```

```python
average_purchases_per_country = average_purchases_per_country.sort_values(by = 'Total_Purchases', ascending = False)
average_purchases_per_country.head(1)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ax1.bar(x = average_purchases_per_country['Country'], 
        height = average_purchases_per_country['Total_Purchases'])
ax1.set_xlabel('Country', fontsize=10)
ax1.set_ylabel('Total_Purchases', fontsize='medium') 

ax2.bar(x = average_purchases_per_country['Country'],
       height = average_purchases_per_country['Purchases_Per_Customer'])
ax2.set_xlabel('Country', fontsize=10)
ax2.set_ylabel('Purchases_Per_Customer', fontsize='medium') 

plt.show()
```

According to the chart, there is no significance with total purchases or purchases per customer in US than other countries so we can accept null hypothesis and reject alternative hypothesis.

## Which products are performing the best, and which are performing the least in terms of revenue?

```python
# Sum the amounts spent in each product category
total_spending = marketing_df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()

# Convert the series to a DataFrame
total_spending_df = total_spending.to_frame(name='Total_Spending')

# Sort the DataFrame in descending order based on total spending
total_spending_df_sorted = total_spending_df.sort_values(by='Total_Spending', ascending=False)

print(total_spending_df_sorted)
```

```python
most_revenue_product = total_spending.idxmax()

most_revenue_amount = total_spending.max()

print(f"The product category that made the most revenue is '{most_revenue_product}' with a total revenue of ${most_revenue_amount:.2f}.")
```

```python
least_revenue_product = total_spending.idxmin()

least_revenue_amount = total_spending.min()

print(f"The product category that made the least revenue is '{least_revenue_product}' with a total revenue of ${least_revenue_amount:.2f}.")
```

## Is there any pattern between the age of customers and the last campaign acceptance rate?


```python
age_groups = marketing_df.groupby('Age')

# Calculate acceptance rates for each age group
acceptance_rates = age_groups[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']].mean()

# Plot acceptance rates against age
plt.figure(figsize=(10, 6))
for campaign in acceptance_rates.columns:
    plt.plot(acceptance_rates.index, acceptance_rates[campaign], label=campaign)

plt.title('Acceptance Rate by Age Group and Campaign')
plt.xlabel('Age')
plt.ylabel('Acceptance Rate')
plt.legend(title='Campaign')
plt.grid(True)
plt.show()
```

```python
age_groups = marketing_df.groupby('Age')

# Calculate acceptance rates for each age group
acceptance_rates = age_groups[['Response']].mean()

# Plot acceptance rates against age
plt.figure(figsize=(10, 6))
for campaign in acceptance_rates.columns:
    plt.plot(acceptance_rates.index, acceptance_rates[campaign], label=campaign)

plt.title('Last Campaign Acceptance Rate by Age Group')
plt.xlabel('Age')
plt.ylabel('Acceptance Rate')
plt.legend(title='Campaign')
plt.grid(True)
plt.show()
```
